## Process lab_data


```python
import re
```

```python
LAB = d_2022['LAB_DATA']
```

```python
def tokenizer(lab):
  #text = re.findall('(Name:\W+)(\w+\s*\w*\s*\w*\s*\w*\s*\w*\s*\w*\s*\w*)(;\W*)',lab)
  text = re.findall('(Name:\W+)(\w+\s?\w*\s?\w*\s?\w*\s?\w*\s?\w*\s?\w*)(;\W*)',lab)
  if text:
    test = text[0][1]
    return test
```

```python
test=[]
for i in range(len(LAB)):
  if LAB[i]==LAB[i]:
    if tokenizer(LAB[i]):
      test.append(tokenizer(LAB[i]))
```

```python
test1 = list(set(test))
len(test1)
```

##Process OTHER_MEDS

```python
OTHER_MEDS = d_2022['OTHER_MEDS']
```

```python
def tok(MED):
  med = re.split(r'[,;\[\]]',MED)
  return med
```

```python
med = []
for i in range(len(OTHER_MEDS)):
  if (OTHER_MEDS[i]==OTHER_MEDS[i] and (OTHER_MEDS[i]!= 'None' and 'none')):
    for j in range(len(tok(OTHER_MEDS[i]))):
      if (len(tok(OTHER_MEDS[i])[j])<=40) and tok(OTHER_MEDS[i])[j]:
        med.append(tok(OTHER_MEDS[i])[j])
```

```python
med1 = list(set(med))
len(med1)
```

## Process CUR_ILL

```python
CUR_ILL = d_2022['CUR_ILL']
```

```python
def tok1(ILL):
  ill0 = re.sub(r'\(.*?\)','',ILL)
  ill = re.split(r'and|[/.,;:]',ill0)
  return ill
```

```python
ill = []
for i in range(len(CUR_ILL)):
  if (CUR_ILL[i]==CUR_ILL[i] and (CUR_ILL[i]!= 'None' and 'none')):
    for j in range(len(tok1(CUR_ILL[i]))):
      if (len(tok1(CUR_ILL[i])[j])<=40) and tok1(CUR_ILL[i])[j]:
        ill.append(tok1(CUR_ILL[i])[j])
```

```python
ill1 = list(set(ill))
len(ill1)
```

## Process ALLERGIES

```python
ALLERGIES = d_2022['ALLERGIES']
```

```python
all = []
for i in range(len(ALLERGIES)):
  if (ALLERGIES[i]==ALLERGIES[i] and (ALLERGIES[i]!= 'None' and 'none')):
    for j in range(len(tok1(ALLERGIES[i]))):
      if (len(tok1(ALLERGIES[i])[j])<=40) and tok1(ALLERGIES[i])[j]:
        all.append(tok1(ALLERGIES[i])[j])
```

```python
all1 = list(set(all))
```

## Process HISTORY

```python
History = d_2022['HISTORY']
```

```python
def token_h(his):
  text = re.findall(r'(Conditions: )(\w+\s*\w*\s*\w*)',his)
  if text:
    dis = text[0][1]
    return dis
```

```python
dis=[]
for i in range(len(History)):
  if History[i]==History[i]:
    if token_h(History[i]):
      dis.append(token_h(History[i]))
```

```python
dis1 = list(set(dis))
len(dis1)
```
